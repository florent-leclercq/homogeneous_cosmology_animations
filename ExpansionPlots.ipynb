```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plot_cubic_lattice(ax, a):
    # Set the mesh size
    size = 3
    
    # Generate lattice points
    x, y, z = np.meshgrid(range(size), range(size), range(size))
    x=a*x+(1-a)
    y=a*y+(1-a)
    z=a*z+(1-a)
    
    # Plot lattice points
    ax.scatter(x, y, z, color='black', marker='o', s=70, depthshade=False)

    # Plot the cubes
    for i in range(size-1):
        for j in range(size):
            for k in range(size):
                ax.plot([x[i,j,k], x[i+1,j,k]], [y[i,j,k], y[i+1,j,k]], [z[i,j,k], z[i+1,j,k]], color='black', lw=2.)
    for i in range(size):
        for j in range(size-1):
            for k in range(size):
                ax.plot([x[i,j,k], x[i,j+1,k]], [y[i,j,k], y[i,j+1,k]], [z[i,j,k], z[i,j+1,k]], color='black', lw=2.)
    for i in range(size):
        for j in range(size):
            for k in range(size-1):
                ax.plot([x[i,j,k], x[i,j,k+1]], [y[i,j,k], y[i,j,k+1]], [z[i,j,k], z[i,j,k+1]], color='black', lw=2.)

    # Remove axis labels
    ax.set_axis_off()

    # Set plot limits
    ax.set_xlim([0, size-1])
    ax.set_ylim([0, size-1])
    ax.set_zlim([0, size-1])

    # Set aspect equal
    ax.set_box_aspect([1, 1, 1])

    # Set projection angle
    ax.view_init(elev=20, azim=35)  # Set the elevation and azimuth angles
```

```python
# Create the subplots
fig = plt.figure(figsize=(4, 4), facecolor='white')

# Plot each subplot
ax = fig.add_subplot(1,1,1, projection='3d')
plot_cubic_lattice(ax, 1.)

# Save figure
fig.tight_layout()
fig.savefig("output/Expansion.png", dpi=300, transparent=True)
```

```python
# Create the subplots
fig = plt.figure(figsize=(10, 2), facecolor='white')
axs = {}
nplots = 5
for i in range(1,nplots+1):
    axs[i] = fig.add_subplot(1,nplots,i, projection='3d')
a = np.array([0.2,0.4,0.6,0.8,1.])

# Plot each subplot
for i in range(1,nplots+1):
    plot_cubic_lattice(axs[i], a[i-1])

# Adjust spacing between subplots
fig.tight_layout(pad=0., w_pad=-0.5)

# Save figure
fig.savefig("output/Expansion_"+str(nplots)+"_panels.png", dpi=300, transparent=True)
```
