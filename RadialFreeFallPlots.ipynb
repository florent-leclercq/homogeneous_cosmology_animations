```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'lines.linewidth': 2})
plt.rcParams.update({'text.usetex': True})
plt.rcParams.update({'text.latex.preamble': r"\usepackage{amsmath}\usepackage{upgreek}"})
plt.rcParams.update({'font.family': 'serif'})
plt.rcParams.update({'font.size': 10})
plt.rcParams.update({'figure.max_open_warning': 20})
```

```python
from RadialFreeFallLib import *
```

```python
v0 = 0.99*vesc
tcrit_1 = tcrit_of_v0(v0)
tmin = 0.
tmax = 2.05*tcrit_1
step = 48*6
N = 5
ts = np.linspace(tmin,tmax,step,endpoint=False)
r_1, v_1, Ec_1, Ep_1, rcrit_1 = integrate(v0, tmin, tmax, step)
r, v, Ec, Ep, rcrit = r_1, v_1, Ec_1, Ep_1, rcrit_1
```

```python
fig, (ax0, ax1, ax2, ax3) = plt.subplots(1,4, figsize=(10,3))
ax0.plot(np.linspace(tmin,tmax,step), r)
ax0.axhline(r0, ls="--", color="black")
if rcrit>0:
    ax0.axhline(rcrit, ls="--", color="black")   
ax0.set_xlabel("$t$")
ax0.set_ylabel("$r(t)$")
ax1.plot(ts, v)
ax1.set_xlabel("$t$")
ax1.set_ylabel("$v(t)$")
ax2.plot(ts, Ec, label="Ec", color="C0")
ax2.plot(ts, Ep, label="Ep", color="C6")
ax2.plot(ts, Ec+Ep, label="Ec+Ep", color="C2")
ax2.set_xlabel("$t$")
ax2.set_ylabel("$E(t)$")
ax2.legend()
ax3.set_ylim([-0.05,1.05])
ax3.plot(ts, Ec/(Ec[0]/N), label="Ec", color="C0")
ax3.plot(ts, -Ep/(Ec[0]/N), label="Ep", color="C6")
fig.tight_layout()
```
